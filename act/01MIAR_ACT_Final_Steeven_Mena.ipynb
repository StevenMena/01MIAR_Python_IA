#Importar librerías

```python
#Dos líbrerias fundamentales en python para el manejo de datos
import numpy as np 
import pandas as pd 

import matplotlib as plt #Librería para plotear o gráficar
import os #Líbreria para manipulación de archivos 
import seaborn as sns #Líbreria que no sirve para crear gráficos o plotear dinámicamente
```

##Cargar dataset
El dataset es un archivo en formato csv.
Para cargar el dataset lo podemos hacer con la librería os o directamente con pandas con su función read_csv.
Para este trabajo se van ocupar dos dataset, que se relacionan por medio del id de la aplicación movil.

```python
#Si nuestro archivo esta en la misma ubicación que este notebook, no necesitamos especificar su folder o la ruta
ruta_dataset = os.path.join('AppleStore.csv')
#Especificamos la ruta del archivo, su carácter como separación, el motor y si posse index propio
#Se almancena el csv en un objeto dataframe de panda
apple_mobile_apps = pd.read_csv(ruta_dataset, sep=',', index_col=0, engine='python')


#Cargando el segundo dataset
#Esta vez demostramos que se puede cargar un dataset en formato csv directamente con la libería pandas
apple_mobile_apps_desc = pd.read_csv('appleStore_description.csv', sep=',', index_col=0, engine='python')
```
